### Insights

- According to the main dataset, 75% of people who engage with the WeRateDogs account prefer to _like_ a tweet instead of retweeting. So perhaps this is something a Social Manager might be interested in or expect from engagement. The dataset shows that the total amount of likes is 107 956 compared to the total amount of retweets which is 56 625. This is a 47% difference between the two variables.
- The _pupper_ stage had the highest number of tweets/dogs in it. The total amount was 176, where the _pupper_ stage made up 51% of the total at 116 dogs.
- The top 3 dogs that had the highest retweets were **Charlie**, **Oliver** and **Lucy**.
